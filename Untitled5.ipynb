```python
import folium

kho_hang = [10.7602, 106.6644]
m = folium.Map(location=kho_hang, zoom_start=12, tiles='CartoDB positron')

folium.Marker(kho_hang, popup="Kho hàng trung tâm", icon=folium.Icon(color='red', icon='home')).add_to(m)

folium.Circle(location=kho_hang,radius=3000,color='green',fill=True,fill_opacity=0.2,popup='Vùng 3km: Giao hàng nhanh < 30p').add_to(m)

folium.Circle(location=kho_hang,radius=5000,color='yellow',fill=True,fill_opacity=0.1,popup='Vùng 5km: Giao hàng trong ngày').add_to(m)


folium.Circle(location=kho_hang,radius=10000,color='red',fill=False, # Không tô màu để đỡ rối mắtpopup='Vùng 10km: Phí giao hàng cao'
).add_to(m)

m.save("BaiTap_ServiceArea.html")
print("Đã tạo bản đồ phân tích vùng phục vụ thành công!")
```
