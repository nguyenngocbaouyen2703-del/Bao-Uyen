```python
import folium
from geopy.distance import geodesic

khach_hang = [10.7602, 106.6644]

danh_sach_xe = {
    "Xe Grab A": [10.7620, 106.6610],
    "Xe Grab B": [10.7550, 106.6700],
    "Xe Grab C": [10.7580, 106.6580]
}

xe_duoc_chon = None
khoang_cach_nho_nhat = float('inf')
for ten_xe, toa_do_xe in danh_sach_xe.items():
    dist = geodesic(khach_hang, toa_do_xe).km
    print(f"Khoảng cách đến {ten_xe}: {dist:.2f} km")

    if dist < khoang_cach_nho_nhat:
        khoang_cach_nho_nhat = dist
        xe_duoc_chon = (ten_xe, toa_do_xe)

print(f"\n---> AI đã chọn: {xe_duoc_chon[0]} (Cách {khoang_cach_nho_nhat:.2f} km)")

m = folium.Map(location=khach_hang, zoom_start=15, tiles='CartoDB positron')

folium.Marker(khach_hang, popup="Khách hàng (Bu)", icon=folium.Icon(color='red', icon='user')).add_to(m)

for ten_xe, toa_do_xe in danh_sach_xe.items():
    color = 'green' if ten_xe == xe_duoc_chon[0] else 'gray'
    folium.Marker(toa_do_xe, popup=ten_xe, icon=folium.Icon(color=color, icon='car')).add_to(m)

folium.PolyLine([khach_hang, xe_duoc_chon[1]], color='blue', weight=3, opacity=0.7).add_to(m)

m.save("BaiTap_GoiXe.html")
```
