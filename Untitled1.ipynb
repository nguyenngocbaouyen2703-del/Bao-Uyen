```python
import folium
from folium import plugins
map_ueh = folium.Map(location = [10.7602, 106.6644], zoom_start = 16)
edu = folium.FeatureGroup(name='truong hoc').add_to(map_ueh)
public = folium.FeatureGroup(name='dia diem cong cong').add_to(map_ueh)

folium.Marker(location = [10.7602, 106.6644], popup='Đại học Kinh tế TP HCM - CS B',tooltip='Click vào ', icon=folium.Icon(color='red')).add_to(edu)
locations = [
    {"name": "Bệnh viện Chợ Rẫy", "coord": [10.7578, 106.6601], "desc": "Bệnh viện đa khoa "},
    {"name": "Bệnh viện Đại học Y Dược", "coord": [10.7554, 106.6653], "desc": "Cơ sở y tế chất lượng "},
    {"name": "The Garden Mall", "coord": [10.7548, 106.6612], "desc": "Trung tâm thương mại giải trí"},
    {"name": "Công viên Văn Lang", "coord": [10.7570, 106.6685], "desc": "Địa điểm vui chơi, chạy bộ "},
    {"name": "Chợ An Đông", "coord": [10.7588, 106.6702], "desc": "Chợ truyền thống "}
]
for loc in locations:
    folium.Marker(location=loc["coord"],popup=folium.Popup(loc["desc"], max_width=200),tooltip=loc["name"],icon=folium.Icon(color='blue', icon='info-sign')).add_to(public)
folium.LayerControl().add_to(map_ueh)
map_ueh.save("BaiTap_Folium.html")
print("Đã tạo bản đồ thành công! Bạn hãy mở file 'BaiTap_Folium.html' để kiểm tra.")





```
