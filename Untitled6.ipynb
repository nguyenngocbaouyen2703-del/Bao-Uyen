```python
import osmnx as ox
import matplotlib.pyplot as plt

vitri = "Quan 10, Ho Chi Minh City, Vietnam"

graph = ox.graph_from_place(vitri, network_type='drive')
fig, ax = ox.plot_graph(graph, node_color='yellow', node_size=10, edge_color= 'white', edge_linewidth=1)

stats = ox.basic_stats(graph)

print(" Kết quả phân tích mạng giao thông {} ".format(vitri))
print("1. Số lượng nút giao (Intersections): {}".format(stats['n']))
print("2. Số lượng đoạn đường (Edges): {}".format(stats['m']))
```
