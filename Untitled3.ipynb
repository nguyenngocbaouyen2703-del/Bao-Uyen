```python
import folium
from folium.plugins import HeatMap
import random


center_hcm = [10.7602, 106.6644]

data_khach_hang = []

for i in range(50):
    lat = center_hcm[0] + random.uniform(-0.01, 0.01)
    lon = center_hcm[1] + random.uniform(-0.01, 0.01)
    data_khach_hang.append([lat, lon])

map_heat = folium.Map(location=center_hcm, zoom_start=15,tiles='CartoDB positron')
HeatMap(data_khach_hang).add_to(map_heat)
map_heat.save("BaiTap_Heatmap.html")
print("Đã tạo xong bản đồ nhiệt! Hãy mở file BaiTap_Heatmap.html")



```
